```python
"""Подключение предустановленных библиотек и собственных модулей"""
import pandas as pd  # as pd - использование псевдонима pd
import math
import time
import random

"""Определение функции"""
def Variables():
    arr = ['a', 'b', 'c'] #список
    print(arr, 'Тип:', type(arr))
    
    a = 3 #переменная
    arr = arr + [a]
    arr.append(math.e) #добавление в конец массива числа e из библиотеки math
    print(*arr)
    arr.clear()
    print("Список пуст:\n", arr)
    
    arr = [i*2 for i in range(10)]
    print("\nСписок из десяти четных чисел:\n", arr)
    
    arr2 = [i for i in range(3, 15, 2)]
    print("\nСписок из чисел [3, 15) с шагом 2:")
    for item in arr2:
        print(item, end = ' ')
        
    
    """Генерация списков"""
    arr2.clear()
    print("\n\nСписок из чисел, которые делятся на 3 (генератор):")
    arr2 = (i for i in range(1, 13) if i % 3 == 0) #Круглые скобки!
    for i in arr2:
        print(i, end = ' ')
    print("\n\nПечать того же генератора: ")
    for i in arr2:
        print(i, end = ' ')
        
Variables() #вызов функции
```

```python
def Dictionary():
    d1 = {'Страна': ['Франция', 'Италия', 'Великобритания'],
              'Столица': ['Париж', 'Рим', 'Лондон']}  #словарь
    print('\n', 'Тип:', type(d1))

    for key, value in d1.items(): 
        print(key, ':', value) #каждый проход цикла печатает 'ключ: значение' словаря   

    """Использование библиотеки pandas"""
    d2 = pd.DataFrame(d1)
    print('\n', 'Использование библиотеки pandas', '\n', d2, '\n')

    """Использование библиотек random и time для создания словаря"""
    d1.clear()
    A = int(time.time()%2)
    B = int(time.time()%13)
    d1 = {key: random.randint(A,B) for key in range(5)}
    print("Словарь с использованием random: ", d1)
    
Dictionary()
```

```python
def Cycles_and_conditions(mas):
    s = 0
    for i in range(len(mas)): #проходим все значения в массиве (списке)
        s = s + mas[i]
    print('Массив:', mas, '\n', 
          'Длина массива:', len(mas), '\n',
         'Сумма всех элементов массива:', s, '\n')
    
    x = 2
    st = '' #строка
    if mas[1] < x < mas[2]:
        st = "It's true, " + str(mas[0]) + ' < ' + str(x) + ' < ' + str(mas[1])
    elif x > max(mas):
        st = "It's true, " + str(x) + ' > ' + str(max(mas))
    else:
        st = "Измените 'x'"
    print(st, '\n')
    
    i = 0
    while True: #бесконечный цикл
        if i == 3: 
            break
        else:
            print(i)
            i = i + 1
    return s, x, st     
        
mas = [1, 2, 3, 4]
print(Cycles_and_conditions(mas))
```

```python
"""Генерация и лямбда-функции"""
def Сreate_Generator():
    mylist = range(4)
    for i in mylist :
        yield i*i #действует как return

mygenerator = Сreate_Generator() # создаём генератор
print("Ошибка:\n", mygenerator) 
for i in mygenerator:
    print(i, end = ' ')
print("\nВторая печать генератора: ")
for i in mygenerator:
    print(i, end = ' ')


print("\nРезультат работы лямбда-функции:", (lambda x: x * 2)(3))

func = lambda x, y: x + y
print("\nРезультат поместили в переменную: ", func(1, 2))
```

### command mode

Чтобы Включить command mode, необходимо Выбрать ячейку и нажать на Esc.
    
    В этом режиме Вам доступны команды по форматированию ноутбука на уровне ячеек:
        1. 'A' - создать ячейку над текущей ячейкой
        2. 'B' - создать ячейку под текущей ячейкой
        3. 'X' - удалить ячейку
        4. 'M' - перевести ячейку в формат "markdown" - это текст на белом фоне
        5. 'Y' - перевести ячейку в формат "Code" - исполняемая ячейка с кодом
        6. 'F' - поиск и замена
        7. shift + M - объединить текущую ячейку со следующей
        8. Больше команд можно увидеть так: Help -> Keyboard shortcuts

### edit mode

В этом режиме можно форматировать ячейки на уровне содержимого.
Для перехода в режим 'edit' из режима 'command' нужно либо кликнуть по ячейки, либо нажать 'enter'.

            * tab - автозаполнение кода или отступ
            * shift + tab - показывает подсказку

```python
fig, ax = plt.subplots()

# можно задать легенду
ax.set_title('Разноцветные кривые')
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')

# добавляем сколько угодно графиков
l1 = ax.plot(x,y1, '--', label = 'sin(x)')
l2 = ax.plot(x,y2, '+:', label = 'cos(x)')
l3 = ax.plot(x,y3, 'h-', label = 'exp(x)')
l4 = ax.plot(x,y4,       label = 'x**2')

#показываем легенду - названия кривых
ax.legend()
```

**Трехмерные графики**

```python
X = np.linspace(-2,2)
Y = np.linspace(-2,2)
x,y = np.meshgrid(X,Y)
t = x**2+y**2
z = np.cos(2*np.pi*t)*np.exp(-t)
```

```python
fig = plt.figure()

ax = fig.add_subplot(projection = '3d')
ax.plot_surface(x,y,z, cmap = 'binary_r')

ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')
```
